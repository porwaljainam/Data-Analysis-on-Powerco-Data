```python
# Extract month from activation date
df_trends["month"] = df_trends["date_activ"].dt.month

# Group by month to detect seasonality
seasonal_trends = df_trends.groupby("month")[["cons_12m", "cons_gas_12m", "cons_last_month"]].mean()

# Define month labels
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
# Plot Electricity Consumption (12M)
plt.figure(figsize=(10, 5))
plt.plot(seasonal_trends.index, seasonal_trends["cons_12m"], marker="o", color="b", label="Electricity Consumption (12M)")
plt.xlabel("Month")
plt.ylabel("Average Consumption")
plt.title("Electricity Consumption (12M) - Seasonal Trend")
plt.xticks(range(1, 13), months)
plt.legend()
plt.grid(True)
plt.show()
```

**interpretation**
 - the electricity consumption is increase in may to july becouse of summer
 - the eletricity consumption is sharp decerse in auguest because of winter

```python
# Plot Gas Consumption (12M)
plt.figure(figsize=(10, 5))
plt.plot(seasonal_trends.index, seasonal_trends["cons_gas_12m"], marker="o", color="r", label="Gas Consumption (12M)")
plt.xlabel("Month")
plt.ylabel("Average Consumption")
plt.title("Gas Consumption (12M) - Seasonal Trend")
plt.xticks(range(1, 13), months)
plt.legend()
plt.grid(True)
plt.show()
```

**interpretation**
 - the gas consumption is increase in november to january becouse of winter
 - the gas consumption is sharp decerse in may to octomber because of summer

# Forecast Analysis

```python
# Function to plot KDE of forecast errors
def plot_kde(actual, forecast, column_name):
    errors = actual - forecast
    
    plt.figure(figsize=(10, 5))
    sns.kdeplot(errors, color='red', fill=True)
    plt.xlabel("Forecast Error (Actual - Forecast)")
    plt.ylabel("Density")
    plt.title(f"Distribution of Forecast Errors: {column_name}")
    plt.grid()
    plt.show()

# Plot KDE for selected columns
plot_kde(df["cons_12m"], df["forecast_cons_12m"], "forecast_cons_12m")
plot_kde(df2["price_peak_var"], df["forecast_price_energy_peak"], "forecast_price_energy_peak")
plot_kde(df2["price_off_peak_fix"], df["forecast_price_pow_off_peak"], "forecast_price_pow_off_peak")
```
